```python
api_key = "aebb0e82-4abd-44d2-a721-abe0f53b1a30"
url_basquet = "http://api.balldontlie.io/v1/players"

```

```python
headers = {
    'Authorization': api_key
}

response = requests.get(url_basquet, headers= headers)
```

```python
def llamar_api(response):
    
    
    print(f"La llamada a la API nos ha dado una respuesta de tipo {response.status_code}")
    
    if response.status_code != 200:
        print(f"El motivo por el que la llamada fallo es {response.reason}")
    else:
        return response.json()
```

```python
llamar_api(response)
```

```python
basquet = response.json()
basquet
```

```python
#necesito iterar en el diccionario y guardar los datos

primer_nombre = []

for nombre in basquet["data"]:
    #controlamos que salga el primer dato
    #print(nombre)
    #iteramos solo en el primer nombre
    #print(nombre["first_name"])
    #ese primer nombre lo ponemos en la lista vacia
    primer_nombre.append(nombre["first_name"])
```

```python
apellido = []

for a in basquet["data"]:
    #controlamos que salga el primer dato
    #print(a)
    #break
    #iteramos solo en el apellido
    #print(a["last_name"])
    #ese primer nombre lo ponemos en la lista vacia
    apellido.append(a["last_name"])
```

```python
equipo = []

for elemento in basquet["data"]:
    #acceder a full name
    diccionario = elemento["team"]["full_name"]
    equipo.append(diccionario)
    
```
