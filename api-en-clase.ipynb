## Selenium

```python
#abrir navegador

driver = webdriver.Chrome()

#acceder a la pagina

driver.get("https://www.bookcenter.es/")

#maximizar ventana

driver.maximize_window()

sleep (3)

#aceptar cookies

driver.find_element("css selector", "#cookie-accept").click()

sleep(5)

#encontrar el buscador el libro de SK

driver.find_element("css selector", "#busqueda > input.search-text").send_keys("Stephen King", Keys.ENTER)


sleep(3)



sleep(2)


libros = []

for i in range (1,4):


#entrar al libro
        driver.find_element("css selector", f"#main-container > div > div > div.row > ul > li:nth-child({i}) > div > div > a > img").click()

#sacar info
        nombre_libro = driver.find_element("css selector", "#titulo").text
        autor = driver.find_element("css selector", "#autor > span").text
        precio_libro = driver.find_element("css selector", "#main-container > div > div.content.col-md-12 > div > div.price-col.col-md-3.col-sm-3 > div.precioDetalle > div.wraper-precioDetalle > div:nth-child(1) > span").text
        precio_descuento = driver.find_element("css selector", "#main-container > div > div.content.col-md-12 > div > div.price-col.col-md-3.col-sm-3 > div.precioDetalle > div.wraper-precioDetalle > div:nth-child(2) > span").text
        editorial = driver.find_element("css selector", "#main-container > div > div.content.col-md-12 > div > div.fichaDetalle.col-md-9.col-sm-9 > div > dl > dd.editorial > span").text
        anio_libro = driver.find_element("css selector", "#main-container > div > div.content.col-md-12 > div > div.fichaDetalle.col-md-9.col-sm-9 > div > dl > dd:nth-child(7)").text
        pagina_libro = driver.find_element("css selector", "#main-container > div > div.content.col-md-12 > div > div.fichaDetalle.col-md-9.col-sm-9 > div > dl > dd:nth-child(14)").text


        libro_nuevo =  {
                "nombre libro": nombre_libro,
                "autor": autor,  
                "precio con descuento": precio_descuento, 
                "precio real": precio_libro,
                "editorial" : editorial,
                "año de edición": anio_libro, 
                "número de páginas": pagina_libro 
            }
 

        libros.append(libro_nuevo)


        sleep(2)
        driver.back()
        sleep(2)



driver.close()
```
